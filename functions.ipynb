## Sigmoid Function

Vamos a agregar una función sigmoide, podemos consultar una función el la siguiente página   [Wolframalpha](https://www.wolframalpha.com/)

Screenshot_20230222_150350.png

```python
def sigmoid(a):      
    return 1 / (1+np.exp(-a))
    #aquí usamos como variable a para evitar usar x que usamos en otros lugares
```

```python
x= np.linspace(10, -10, 100)
#Aquí generamos una iteración para que me traiga 100 numeros
#del 10 al -10
```

```python
x
```

## Generar una visualización de los valores

```python
plt.plot(x,sigmoid(x))
```

## Función Step (escalonada,paso)

Screenshot_20230222_151336.png

```python
def step(x):      
    return np.piecewise(x,[x<0.0, x>0.0], [0,1])
    ## piecewise nos ayuda para condiciones
```

```python
plt.plot(x,step(x))
```

## Función Relu

Screenshot_20230222_161853.png

```python
def relu(x):
  return np.piecewise(x, [x < 0, x > 0], [0, lambda x: x])
```

```python
plt.plot(x,relu(x))
```

## Funcion de perdida (loss function)

MSE-2bf0c267-2166-4071-bcde-d0e9589a5dc6.webp

```python
def mse(y, y_hat, derivate=False):
  if derivate:
      return (y_hat -y)
  else:
      return np.mean((y_hat -y) **2)
```

```python
real = np.array([0,0,1,1])
prediction = np.array([0.9,0.5,0.2,0.0])
```

```python
mse(real, prediction)
```
